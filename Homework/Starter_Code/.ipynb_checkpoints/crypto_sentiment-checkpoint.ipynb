# Tokenizer

In this section, you will use NLTK and Python to tokenize the text for each coin. Be sure to:
1. Lowercase each word
2. Remove Punctuation
3. Remove Stopwords

```python
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from string import punctuation
import re
```

```python
# Expand the default stopwords list if necessary
# YOUR CODE HERE!
sw_addon = {'chars'}
```

```python
# Complete the tokenizer function
def tokenizer(text, sw_addon = {}):
    """Tokenizes text."""
    #standard stopwords
    sw = set(stopwords.words('english'))
    #clean punctuation
    regex = re.compile("[^a-zA-Z ]")
    re_clean = regex.sub('', text)
    #tokenize words
    words = word_tokenize(re_clean)
    #Lemmatize all words
    lem = [WordNetLemmatizer().lemmatize(word) for word in words]
    #negative union words with all stopwords
    tokens = [word.lower() for word in lem if word.lower() not in sw.union(sw_addon)]
    return tokens

```

```python
# Create a new tokens column for bitcoin
# YOUR CODE HERE!

bitcoin_sentiment_df["Tokens"] = bitcoin_sentiment_df["Text"].apply(tokenizer)
bitcoin_sentiment_df.head()
```

```python
# Create a new tokens column for ethereum
# YOUR CODE HERE!
eth_sentiment_df["Tokens"] = eth_sentiment_df["Text"].apply(tokenizer)
eth_sentiment_df.head()
```

# NGrams and Frequency Analysis

In this section you will look at the ngrams and word frequency for each coin. 

1. Use NLTK to produce the n-grams for N = 2. 
2. List the top 10 words for each coin. 

```python
from collections import Counter
from nltk import ngrams
```

```python
#bitcoin_sentiment_df["Tokens"].head()

def get_tokens(df):
    tokens = []
    for i in df:
        tokens.extend(i)
    return tokens


#all words
btc_tokens = get_tokens(bitcoin_sentiment_df["Tokens"])
eth_tokens = get_tokens(eth_sentiment_df["Tokens"])


```

```python
#words with count
btc_word_count = Counter(btc_tokens)
eth_word_count = Counter(eth_tokens)

btc_word_count['cryptocurrency']
```

```python
# Generate the Bitcoin N-grams where N=2
# YOUR CODE HERE!


def get_token(tokens, N):
    ngram_counts = Counter(ngrams(tokens, N))
    return ngram_counts

btc_bigrams = get_token(btc_tokens,2)
```

```python
# Generate the Ethereum N-grams where N=2
# YOUR CODE HERE!

eth_bigrams = get_token(eth_tokens,2)
```

```python
# Use the token_count function to generate the top 10 words from each coin
def token_count(tokens, N=10):
    """Returns the top N tokens from the frequency count"""
    return Counter(tokens).most_common(N)
```

```python
# Get the top 10 words for Bitcoin
# YOUR CODE HERE!
top_10_bitcoin = token_count(btc_tokens,10)
top_10_bitcoin
```

```python
# Get the top 10 words for Ethereum
# YOUR CODE HERE!
top_ten_eth = token_count(eth_tokens,10)
top_ten_eth
```

# Named Entity Recognition

In this section, you will build a named entity recognition model for both coins and visualize the tags using SpaCy.

```python
import spacy
from spacy import displacy
```

```python
# Optional - download a language model for SpaCy
!python -m spacy download en_core_web_sm
```

```python
# Load the spaCy model
nlp = spacy.load('en_core_web_sm')
```

## Bitcoin NER

```python
# Concatenate all of the bitcoin text together
# YOUR CODE HERE!
btc_content = ''.join(bitcoin_sentiment_df["Text"])
```

```python
# Run the NER processor on all of the text
# YOUR CODE HERE!
btc_doc = nlp(btc_content)
# Add a title to the document
# YOUR CODE HERE!

btc_doc.user_data["title"] = "Bitcoin NER Content"
```

```python
# Render the visualization
# YOUR CODE HERE!
displacy.render(btc_doc, style = 'ent')
```

```python
# List all Entities
# YOUR CODE HERE!
for ent in btc_doc.ents:
    print('{} {}'.format(ent.text, ent.label_))
    
```
